```python
# Splitting the image Colors

Lake_img = cv2.imread('Lake.jpg', cv2.IMREAD_COLOR)
B, G, R = cv2.split(Lake_img)

# Showing the channels
plt.figure(figsize = [20, 5])
plt.subplot(141);
plt.imshow(R, cmap = 'gray');
plt.title("Red-Channel");

plt.subplot(142);
plt.imshow(G, cmap = 'gray');
plt.title("Green-Channel");

plt.subplot(143);
plt.imshow(B, cmap = 'gray');
plt.title("Blue-Channel");

# Merging the Color Chanenl

merged_img = cv2.merge((B,G,R))
plt.subplot(144);
plt.imshow(merged_img[:,:,::-1]);
plt.title("Merged Image");
```

Converting in different color Spaces

```python
Lake_img = cv2.cvtColor(Lake_img, cv2.COLOR_RGB2BGR)
plt.imshow(Lake_img)
```

```python
#Changing to HSV Color

Img_Hsv = cv2.cvtColor(Lake_img,cv2.COLOR_BGR2RGB)
Img_Hsv = cv2.cvtColor(Lake_img,cv2.COLOR_RGB2HSV)

h, s, v = cv2.split(Img_Hsv)

plt.figure(figsize=[25,8])
plt.subplot(141);plt.imshow(h);plt.title("H Channel");
plt.subplot(142);plt.imshow(s);plt.title("S Channel");
plt.subplot(143);plt.imshow(v);plt.title("V Channel");
plt.subplot(144);plt.imshow(Img_Hsv);plt.title("Original Picture");
```

```python
# Modifying Individual Channel

h_new = h + 20
Lake_Merged = cv2.merge((h_new, s, v))
new_img = cv2.cvtColor(Lake_Merged, cv2.COLOR_HSV2RGB)

plt.figure(figsize=[25,8])
plt.subplot(141);plt.imshow(h_new);plt.title("H manipulated Channel");
plt.subplot(142);plt.imshow(s);plt.title("S Channel");
plt.subplot(143);plt.imshow(v);plt.title("V Channel");
plt.subplot(144);plt.imshow(new_img);plt.title("Manipulated image");
```

```python
cv2.imwrite('LakeSaved.png',Lake_Merged)
Image(filename='LakeSaved.png')
```
