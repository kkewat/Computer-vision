```python
img2 = cv2.imread('mount2.jpg',cv2.IMREAD_COLOR)
img2_rgb = img2[:,:,::-1]
plt.imshow(img2_rgb)
```

```python
img_crop = img2_rgb[200:400,400:650]
plt.imshow(img_crop)
```

```python
img_resize = cv2.resize(img_crop,None, fx=2, fy=2)
plt.imshow(img_resize)
```

```python
d_width = 100
d_height = 200
dim = (d_width,d_height)

croped_resize = cv2.resize(img_crop, dsize=dim, interpolation = cv2.INTER_AREA)
plt.imshow(croped_resize)
```

```python
desired_width = 100
aspect_ratio = desired_width / img_crop.shape[1]
desired_height = int(img_crop.shape[0] * aspect_ratio)
dim = (desired_width, desired_height)

# Resize image
resized_cropped_region = cv2.resize(img_crop, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
```

```python
img_crop = img_crop[:,:,::-1]
cv2.imwrite('cropped_image.jpg',img_crop)
Image(filename='cropped_image.jpg')
```

```python
flip_horizontal = cv2.flip(img_crop,1)
flip_vertical = cv2.flip(img_crop,0)
flip_both =cv2.flip(img_crop,-1)

plt.figure(figsize=(18,5))
plt.subplot(141);plt.imshow(flip_horizontal);
plt.subplot(142);plt.imshow(flip_vertical);
plt.subplot(143);plt.imshow(flip_both);
plt.subplot(144);plt.imshow(img_crop);plt.title('original');
```
